```python
#1
def fix_names_and_marital_status(data):
    for student in data.values():
        student["first name"] = student["first name"].capitalize()
        student["last name"] = student["last name"].capitalize()
        student["marital status"] = student["marital status"].lower()
fix_names_and_marital_status(Golestan_System)
print(Golestan_System)
```

```python
#1_2
def fix_residence(data):
    for student in data.values():
        if student["Place of resident"] == "":
            student["Place of resident"] = "Tehran"
fix_residence(Golestan_System)
print(Golestan_System)            
```

```python
#1_3
def update_ages(golestan_system, current_year=2024):
    for student in golestan_system.values():
        birth_year = student["age"]
        real_age = current_year - birth_year
        student["age"] = real_age
update_ages(Golestan_System)
print(Golestan_System)
```

```python
#1_4
def fix_student_ids(golestan_system):
    for student in golestan_system.values():
        student["stu_num"], student["National Code"] = student["National Code"], student["stu_num"]
fix_student_ids(Golestan_System)
print(Golestan_System)
```

```python
#2
import matplotlib.pyplot as plt

def plot_scores(golestan_system):
    students = [student["first name"] for student in golestan_system.values()]
    comm_scores = [student["lessons"]["Communication system"] for student in golestan_system.values()]
    num_scores = [student["lessons"]["Numerical calculations"] for student in golestan_system.values()]
    
    avg_comm = sum(comm_scores) / len(comm_scores)
    avg_num = sum(num_scores) / len(num_scores)
    plt.plot(students, comm_scores, color='blue', marker='o', label='Communication System')
    plt.plot(students, num_scores, color='red', marker='o', label='Numerical Calculations')
    
    plt.axhline(avg_comm, color='blue', linestyle='--', label=f'Avg Comm: {avg_comm:.1f}')
    plt.axhline(avg_num, color='red', linestyle='--', label=f'Avg Num: {avg_num:.1f}')
    
    plt.title('Student Scores: Communication & Numerical')
    plt.xlabel('Students')
    plt.ylabel('Scores')
    plt.grid(True)
    plt.show()
plot_scores(Golestan_System)
```

```python
#3
import matplotlib.pyplot as plt

def plot_grade_distribution(golestan_system):
    all_scores = []
    for student in golestan_system.values():
        all_scores.extend(student["lessons"].values())
    grade_bins = {'A': 0, 'B': 0, 'C': 0, 'D': 0}
    for score in all_scores:
        if 17 < score <= 20:
            grade_bins['A'] += 1
        elif 14 <= score < 17:
            grade_bins['B'] += 1
        elif 10 <= score < 14:
            grade_bins['C'] += 1
        elif 0 <= score < 10:
            grade_bins['D'] += 1
    labels = grade_bins.keys()
    sizes = grade_bins.values()
    colors = ['#4CAF50', '#2196F3', '#FFC107', '#F44336'] 

    plt.figure(figsize=(7, 7))
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
    plt.title('Grade Distribution of All Scores [University of Isfahan]')
    plt.axis('equal') 
    plt.show()
plot_grade_distribution(Golestan_System)
```
